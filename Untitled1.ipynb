```python
##Construir un algoritmo de red neuronal artificial con el perceptrón “AND”##
```

```python
from sklearn.linear_model import Perceptron


X = [[0, 0], [0, 1], [1,0], [1,1]]
y = [0, 0, 0, 1]

perceptron = Perceptron()

perceptron.fit(X, y)

predicciones = perceptron.predict(X)

for entrada, prediccion in zip(X, predicciones):
  print(f'Entrada: {entrada} - Prediccion: {prediccion}')
```

Construir un algoritmo de una red neuronal artificial

```python
import numpy as np
import keras
from keras.models import Sequential
from keras.layers import Dense

x_train = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_train = np.array([0, 0, 0, 1])

model = Sequential()
model.add(Dense(2, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x_train, y_train, epochs=100, batch_size=1)

scores = model.evaluate(x_train, y_train)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

x_test = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
predictions = model.predict(x_test)
rounded = [round(x[0]) for x in predictions]
print(rounded)
```
